**Titanic Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
titanic_data=pd.read_csv("train_and_test2.csv")
```

```python
titanic_data
```

```python
# Display the first few row
print(titanic_data.head())
```

```python
# Get basic information about the dataset
print(titanic_data.info())
```

```python
# Get summary statistics
print(titanic_data.describe())
```

```python
# Check for missing values
print(titanic_data.isnull().sum())
```

```python
# For simplicity, let's fill missing values for 'Age' with the median age and 'Embarked' with the mode
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
# Drop the 'Cabin' column since it has too many missing values
titanic_data.drop(columns=['Embarked'], inplace=True)
```

```python
# Verify that there are no more missing values
print(titanic_data.isnull().sum())
```

```python
# Create age groups
bins = [0, 12, 20, 40, 60, 80]
labels = ['Child', 'Teenager', 'Adult', 'Middle-Aged', 'Senior']
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=bins, labels=labels, right=False)

# Survival rate by age group
age_group_survival = titanic_data.groupby('AgeGroup')['2urvived'].mean()
age_group_survival.plot(kind='bar', color='pink')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Distribution of 'Survived'
titanic_data['2urvived'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

```

```python
# Distribution of 'Age'
plt.hist(titanic_data['Age'], bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Survival rate by gender
titanic_data.groupby('Sex')['2urvived'].mean().plot(kind='bar', color=['pink', 'green'])
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
x=titanic_data['Age']
y=titanic_data['Fare']
plt.scatter(x,y,c="purple")
```

```python
# Compute the correlation matrix
corr_matrix = titanic_data.corr()
```

```python
# Plot the correlation matrix
plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```
